###4) What does a P-value represent in hypothesis testing?

->

* A p-value in hypothesis testing represents the probability of obtaining the observed results, or more extreme results, assuming that the null hypothesis is true.
* It helps measure the strength of evidence against the null hypothesis.
* A smaller p-value indicates stronger evidence to reject the null hypothesis.  
   * For example, a p-value of 0.03 means there's a 3% chance that the observed outcome could have occurred under the assumption that the null hypothesis is true. If the p-value is less than the chosen significance level (like 0.05), the null hypothesis is rejected.

###12) What is the T-test, and how is it used in hypothesis testing?

->

* A T-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between means, especially when the sample size is small and the population standard deviation is unknown.
* It uses the t-distribution to evaluate the difference between a sample mean and a population mean or between two sample means.
* In hypothesis testing, the t-test helps decide whether to reject the null hypothesis by calculating a t-value, which is then compared to a critical value or used to find the p-value.
* If the p-value is less than the chosen significance level (e.g., 0.05), the null hypothesis is rejected, indicating a statistically significant difference.

---

###13) What is the relationship between Z-test and T-test in hypothesis testing?

->

The Z-test and T-test are both statistical methods used in hypothesis testing to compare sample data with a population or between two samples. They are similar in purpose but differ in application based on the data characteristics.

***Relationship:***

* Both tests assess whether differences between means are statistically significant.

* They use similar formulas to calculate test statistics, comparing the sample mean to a known or hypothesized value.

* The Z-test is used when the population standard deviation is known and the sample size is large.

* The T-test is used when the population standard deviation is unknown and/or the sample size is small.

* As the sample size increases, the t-distribution approaches the normal distribution, making the t-test results very similar to the Z-test.

***In short:***

Both tests serve the same goal in hypothesis testing, but the Z-test is for known standard deviation and large samples, while the T-test is for unknown standard deviation and small samples.

###14) What is a confidence interval, and how is it used to interpret statistical results?

->

A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence ‚Äî typically 95% or 99%.

***How it's used:***

* A 95% confidence interval means that if we repeated the sampling process many times, about 95% of the calculated intervals would contain the true population value.

* It gives both an estimate and the uncertainty around that estimate.

* In hypothesis testing, if the confidence interval does not include the value stated in the null hypothesis (e.g., a mean of 0), it suggests the result is statistically significant.

***Example:***

If a 95% CI for the average test score is (72, 78), we are 95% confident that the true average score lies between 72 and 78.

***In short:***

 a confidence interval shows the likely range of a population value and helps determine the reliability of your estimate.

---

###15)  What is the margin of error, and how does it affect the confidence interval?

->

The margin of error is the amount added and subtracted from the sample estimate to create a confidence interval. It reflects the maximum expected difference between the sample statistic and the true population parameter due to sampling variability.

***How it affects the confidence interval:***

* The larger the margin of error, the wider the confidence interval ‚Äî indicating less precision.

* The smaller the margin of error, the narrower the interval ‚Äî indicating more precision.

* It depends on the sample size, confidence level, and data variability. A larger sample size or lower confidence level will reduce the margin of error.

***In short:***

The margin of error controls how wide the confidence interval is, directly affecting how precisely you can estimate the population value.

###19) What is the F-distribution, and when is it used in hypothesis testing?

->

The F-distribution is a probability distribution used primarily in hypothesis testing to compare two variances. It is right-skewed and only takes positive values.

***When is it used:***

* In the ANOVA (Analysis of Variance) test to compare the means of three or more groups.

* To test if two population variances are significantly different.

* In regression analysis, to evaluate the overall significance of a model.

***In short:***

The F-distribution is used in hypothesis testing when comparing variances or testing multiple group means, especially in ANOVA and regression analysis.

---

###20) A What is an ANOVA test, and what are its assumptions?

->

ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more groups. It uses the F-distribution to compare the variability between group means to the variability within the groups.

***Key Assumptions of ANOVA:***

**Independence:** Observations are independent of each other.

**Normality:** The data in each group is approximately normally distributed.

**Homogeneity of variances:** All groups have similar variances (equal spread).

***In short:***

ANOVA tests if at least one group mean differs significantly from the others, assuming the data is independent, normally distributed, and has equal variances.


---

###21) A What are the different types of ANOVA tests?

->

There are three main types of ANOVA tests, each used based on the structure of the data and the number of factors (independent variables) involved:

***One-Way ANOVA:***

Used to compare the means of three or more groups based on one independent variable.

Example: Testing if average test scores differ across different teaching methods.

***Two-Way ANOVA:***

Used to compare means based on two independent variables and also checks for interaction effects between them.

Example: Studying the effect of diet and exercise on weight loss.

***Repeated Measures ANOVA:***

Used when the same subjects are measured multiple times under different conditions or over time.

Example: Measuring student performance before, during, and after a training program.

***In short:***

One-Way ANOVA: One factor, multiple groups

Two-Way ANOVA: Two factors, checks interaction

Repeated Measures ANOVA: Same subjects, repeated testing over time or conditions

```python
#2)  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for the discrete uniform distribution
low = 1   # inclusive lower bound
high = 7  # exclusive upper bound

# Create a discrete uniform distribution
dist = randint(low, high)

# Generate x values (possible outcomes)
x = np.arange(low, high)

# Compute PMF for each outcome
pmf_values = dist.pmf(x)

# Plot the PMF
markerline, stemlines, baseline = plt.stem(x, pmf_values, basefmt=" ")
plt.setp(markerline, marker='o', markersize=6, color='blue')
plt.setp(stemlines, linestyle='-', color='blue')
plt.xlabel('Outcome')
plt.ylabel('PMF')
plt.title('PMF of Discrete Uniform Distribution')
plt.grid(True)
plt.show()
```

```python
#3)  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome, either 0 or 1
    p (float): Probability of success (1), where 0 ‚â§ p ‚â§ 1

    Returns:
    float: PMF value for the given x
    """
    if x not in [0, 1]:
        return 0.0  # PMF is 0 for values not in {0, 1}
    return p if x == 1 else 1 - p

# Example usage:
print("P(X=0):", bernoulli_pmf(0, 0.7))
print("P(X=1):", bernoulli_pmf(1, 0.7))
```

```python
#9) Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for the dataset and plot them.

    Parameters:
    data (list or numpy array): Input numeric dataset
    """
    # Convert to numpy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.scatter(range(len(z_scores)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Z = 0')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of Dataset')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage:
sample_data = [10, 12, 9, 14, 13, 15, 11, 8, 9, 13]
z_scores = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z_scores)
```

```python
#5)  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)
    x_crit_low = mu0 - z_crit * se
    x_crit_high = mu0 + z_crit * se

    # X values for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Null and alternative distributions
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(12, 6))
    plt.plot(x, y0, label='H‚ÇÄ Distribution (Œº = {})'.format(mu0), color='blue')
    plt.plot(x, y1, label='H‚ÇÅ Distribution (Œº = {})'.format(mu1), color='green')

    # Shade Type I Error regions (rejection of H0 when H0 is true)
    plt.fill_between(x, y0, where=(x < x_crit_low) | (x > x_crit_high), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Shade Type II Error region (fail to reject H0 when H1 is true)
    plt.fill_between(x, y1, where=(x > x_crit_low) & (x < x_crit_high), color='orange', alpha=0.4, label='Type II Error (Œ≤)')

    # Vertical lines at critical values
    plt.axvline(x_crit_low, color='black', linestyle='--')
    plt.axvline(x_crit_high, color='black', linestyle='--')

    # Labels and title
    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_type1_type2_errors()
```

```python
#11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):
    # Calculate the complement of the hypothesis
    prior_not_H = 1 - prior_H

    # Total probability of data (evidence)
    evidence = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)

    # Apply Bayes' Theorem
    posterior = (likelihood_D_given_H * prior_H) / evidence
    return posterior
```

```python
# Given:
prior_H = 0.01                   # P(H): Disease probability
likelihood_D_given_H = 0.99      # P(D|H): True positive rate
likelihood_D_given_not_H = 0.05  # P(D|¬¨H): False positive rate

# Calculate posterior probability
posterior = bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
#16)  Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

# Step 1: Generate two independent random samples
np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=50)  # mean=10, std=2
sample2 = np.random.normal(loc=10, scale=3, size=50)  # mean=10, std=3

# Step 2: Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: Compute F-statistic
F_statistic = var1 / var2 if var1 > var2 else var2 / var1

# Step 4: Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Step 5: Compute p-value (two-tailed)
p_value = 2 * min(f.cdf(F_statistic, df1, df2), 1 - f.cdf(F_statistic, df1, df2))

# Step 6: Display the results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#19) Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):
    # Flatten all groups for combined data
    all_data = np.concatenate(groups)

    print("\n--- Normality Test (Shapiro-Wilk) ---")
    for i, group in enumerate(groups, start=1):
        stat, p = stats.shapiro(group)
        print(f"Group {i}: W = {stat:.4f}, p-value = {p:.4f} ‚Üí {'Normal' if p > 0.05 else 'Not normal'}")

    print("\n--- Homogeneity of Variances (Levene's Test) ---")
    stat, p = stats.levene(*groups)
    print(f"Levene‚Äôs stat = {stat:.4f}, p-value = {p:.4f} ‚Üí {'Equal variances' if p > 0.05 else 'Unequal variances'}")

    print("\n--- Boxplot for Visual Inspection ---")
    # Plot boxplots for visual variance check
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=groups)
    plt.xlabel("Group")
    plt.ylabel("Values")
    plt.title("Boxplot of Groups")
    plt.grid(True)
    plt.show()

    print("\n--- QQ Plots for Normality Check ---")
    # QQ plots for normality
    for i, group in enumerate(groups, start=1):
        plt.figure()
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"QQ Plot - Group {i}")
        plt.grid(True)
        plt.show()

    print("\n* Note: Independence is assumed from experimental design.")

```

```python
#20)  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Simulate data
np.random.seed(42)

# Define factors
gender = ['Male', 'Female']
method = ['A', 'B']

# Create a full factorial design
data = []
for g in gender:
    for m in method:
        # Simulate scores with different means for interaction
        mean = 75 if g == 'Male' and m == 'A' else 80 if g == 'Male' else 78 if m == 'A' else 82
        scores = np.random.normal(loc=mean, scale=5, size=30)
        for score in scores:
            data.append([g, m, score])

df = pd.DataFrame(data, columns=['Gender', 'Method', 'Score'])

# Step 2: Two-way ANOVA using statsmodels
model = ols('Score ~ C(Gender) + C(Method) + C(Gender):C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\n--- Two-Way ANOVA Table ---")
print(anova_table)

# Step 3: Visualization with interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x='Method', y='Score', hue='Gender', capsize=0.1, errwidth=1.5)
plt.title('Interaction Plot: Method vs Score by Gender')
plt.grid(True)
plt.show()
```

```python
#24) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy.stats import chi2

# Step 1: Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=50, scale=5, size=30)  # population std dev = 5
n = len(sample)
sample_var = np.var(sample, ddof=1)

# Step 2: Hypothesized population variance
sigma_0_squared = 25  # (std dev = 5, same as above)

# Step 3: Compute Chi-square test statistic
chi2_stat = (n - 1) * sample_var / sigma_0_squared

# Step 4: Determine critical values for 95% confidence level
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Step 5: Print results
print("=== Chi-Square Test for Population Variance ===")
print(f"Sample Variance: {sample_var:.2f}")
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"Chi-square Critical Values: [{chi2_lower:.2f}, {chi2_upper:.2f}]")

# Step 6: Interpretation
if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print("Conclusion: Reject the null hypothesis (variance is significantly different).")
else:
    print("Conclusion: Fail to reject the null hypothesis (no significant difference in variance).")
```

```python
#26)  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate two datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=100)   # std = 5
data2 = np.random.normal(loc=55, scale=10, size=100)  # std = 10

# Step 2: Calculate variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Calculate F-statistic (larger variance on top)
F = var2 / var1 if var2 > var1 else var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 4: Calculate critical value for 95% confidence
alpha = 0.05
f_critical = f.ppf(1 - alpha/2, df1, df2)

# Step 5: Print results
print("=== F-Test for Comparing Variances ===")
print(f"Variance 1: {var1:.2f}")
print(f"Variance 2: {var2:.2f}")
print(f"F-statistic: {F:.4f}")
print(f"F-critical (two-tailed, alpha=0.05): {f_critical:.4f}")

if F > f_critical:
    print("Conclusion: Reject the null hypothesis (variances are significantly different).")
else:
    print("Conclusion: Fail to reject the null hypothesis (no significant difference in variances).")

# Step 6: Visualize distributions of datasets
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(data1, kde=True, color='skyblue', label='Dataset 1')
sns.histplot(data2, kde=True, color='salmon', label='Dataset 2')
plt.title("Distributions of Two Datasets")
plt.legend()
plt.grid(True)

# Step 7: Visualize F-distribution with the test statistic
plt.subplot(1, 2, 2)
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, df1, df2)
plt.plot(x, y, 'b-', label='F-distribution')
plt.axvline(F, color='red', linestyle='--', label=f'F-stat = {F:.2f}')
plt.axvline(f_critical, color='green', linestyle='--', label=f'Critical value = {f_critical:.2f}')
plt.fill_between(x, y, where=(x > f_critical), color='orange', alpha=0.3)
plt.title("F-Distribution and Decision Boundary")
plt.xlabel('F-value')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```
